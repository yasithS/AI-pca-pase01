### Function

```python
# prompt: i want to create a function to select multiple columns from a excel and if column name is country, pie chart should popup otherwise a normal bar chart

import matplotlib.pyplot as plt
def plot_columns(dataframe, column_names):
  """
  Selects multiple columns from a dataframe and plots them.
  If a column name is 'Country', a pie chart is generated.
  Otherwise, a bar chart is generated.

  Args:
    dataframe: The pandas DataFrame.
    column_names: A list of column names to plot.
  """
  for col_name in column_names:
    if col_name not in dataframe.columns:
      print(f"Warning: Column '{col_name}' not found in the DataFrame.")
      continue

    data = dataframe[col_name].value_counts()
    x = data.index.tolist()
    y = data.values.tolist()

    fig, ax = plt.subplots()

    if col_name.lower() == 'country':
      ax.pie(y, labels=x, autopct='%1.1f%%')
      ax.set_title(f'Distribution of {col_name}')
    else:
      ax.bar(x, y)
      ax.set_xlabel(col_name)
      ax.set_ylabel('Count')
      ax.set_title(f'Count of values in {col_name}')
      plt.xticks(rotation=90)

    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming 'df' is your DataFrame loaded from the Excel file
# plot_columns(df, ['Country', 'City', 'Age'])
```

```python
plot_columns(df, ['Country', 'Job Title'])
```
